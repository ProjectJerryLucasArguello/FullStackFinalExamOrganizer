```python
#We need to remove all these Nan as they were once separated by a line of commas
df_clean =df.dropna(how = 'all')
df_clean
```

```python
df_copy = df_clean.copy()
```

```python
df_copy['Course ID']
```

```python
#I need to convert this into a string so I can split it
df_copy['Course ID'] = df_copy['Course ID'].astype(str)
```

```python
df_copy['Course ID']
```

```python
#df_copy['Course ID'].str.extract(r'([A-Za-z]+)(\d+)')
```

```python
#I have split the Course ID into the two important components: the class and class_level
# s? accounts for either a single or no space | s* accounts for multiple space or no spaces
df_copy[['class', 'class_level']] = df_copy['Course ID'].str.extract(r'([A-Za-z]{2,4})\s*(\d{3})')
df_copy
```

```python
#So here I am going to delete the Course ID column and the Exam Component ID
df_copy = df_copy.drop(['Course ID'], axis = 1)
df_copy
```

```python
#df_copy.to_csv('df_finalExamsOrganizer')
```

```python
#For the purpose of avoidng future complications I shall fill the columns in the Exam Component ID as three digits
df_copy['Exam Component ID'] = df_copy['Exam Component ID'].astype(str).str.zfill(3)
df_copy
```
